```python
plt.figure(figsize=(10, 5))
plt.scatter(crime_data1.State, crime_data1.Murder, color='r')
plt.scatter(crime_data1.State, crime_data1.Assault, color='g')
plt.scatter(crime_data1.State, crime_data1.UrbanPop, color='b')
plt.scatter(crime_data1.State, crime_data1.Rape, color='y')
plt.xlabel('state')
plt.ylabel('rate')
plt.show()
```

```python
def normalf(i):
    x = (i - i.min())/(i.max()-i.min())
    return(x)
```

```python
crime_data_normal = normalf(crime_data1.iloc[:,1:])
```

```python
crime_data_normal
```

```python
#plotting dendrogram
plt.figure(figsize=(12,6))
dendrogram = sch.dendrogram(sch.linkage(crime_data_normal,method='ward'))
```

```python
#Fitting & Predicting
#specifying hierarchical clustering, clusters as 3(we are concidering n of cluster =3)
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
hc
```

```python
#prediction - there is no centroid in agglomerative options
y_hc = hc.fit_predict(crime_data_normal)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
crime_data1['Hierarchical_clusterid'] = Clusters
```

```python
crime_data1
```

```python
crime_data1.groupby('Hierarchical_clusterid').agg(['mean']).reset_index()
```

```python
crime_data1['Hierarchical_clusterid'].value_counts()
```

```python
#plotting clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime_data1['Hierarchical_clusterid'],crime_data1['Murder'], c=hc.labels_) 
```

```python
# 3 clusters is formed
```

# **Kmeans**

```python
#import necessary libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_crime_data = scaler.fit_transform(crime_data.iloc[:,1:])
```

```python
scaled_crime_data
```

```python
#Identify the best k-value using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_data)
    wcss.append(kmeans.inertia_) 
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#k = 4 is the best we can see
#Build Cluster algorithm
clusters_new = KMeans(4, random_state=42) #randomly divide data into 4 partitions. Use any number but keep consistent
clusters_new
```

```python
clusters_new.fit(scaled_crime_data)
clusters_new.labels_
```

```python
crime_data1['ClusterID_new']=clusters_new.labels_
crime_data1.head()
```

```python
#Average values of all data points within one cluster for profiling purposes.
clusters_new.cluster_centers_
```

```python
crime_data1.groupby('ClusterID_new').agg(['mean']).reset_index()
```

```python
crime_data1['ClusterID_new'].value_counts()
```

```python
plt.scatter(crime_data1['ClusterID_new'],crime_data['Murder'],c = clusters_new.labels_)
```
